```python
import pandas as pd 
df_test = pd.read_csv('test.csv')
df_test
```

```python
df_labels = pd.read_csv('test_labels.csv')
df_labels
```

```python
df = pd.merge(df_test, df_labels, left_index=True, right_index=True) # merge both dataframes tgt
df = df.drop(df.columns[0], axis=1) # removing ids
df = df.drop(df.columns[1], axis=1)
df
# according to source,
# "Since the data was originally used for a Kaggle competition, 
# in the test_labels dataset there are observations with labels of value -1 indicating it was not used for scoring."
```

```python
# making dataframe into appropriate format
lst = []
for i in range(len(df)):
    values = df['toxic'][i], df['severe_toxic'][i], df['obscene'][i], df['threat'][i], df['insult'][i], df['identity_hate'][i]
    lst.append(values)
    
#print(lst[:5])

dic = {0:'toxic', 1:'severe_toxic', 2:'obscene', 3:'threat', 4:'insult', 5:'identity_hate'}
result = []
for row in lst:
    temp = ""
    for i in range(len(row)):
        if row[i] == 1:
            temp += dic[i] + "|"
        elif row[i] == -1:
            temp = "-1."
        else:
            continue
        
    temp = temp[:-1]
    if temp == "":
        temp = "non_hate"
        
    result.append(temp)
    
#print(result[:5])
```

```python
# removing "\n" from comment_text
new_text = []
for i in range(len(df)):
    new_text.append(df['comment_text'][i].replace("\n", " "))
    
df = df.drop("comment_text", axis=1)
df.insert(0, "comment_text", new_text)
    
df
```

```python
# relabelling dataframe
#df = df.drop("label", axis=1)
df.insert(0, "label", result)
df
```

```python
# only label and comment_text cols
df = df[["label","comment_text"]]
df
```

```python
# remove rows with labels -1
df = df.drop(df[df.label == '-1'].index)
df
```
